# Model Building

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import plot_model

def create_model(hidden_layers = 2, neurons = 20, activation = 'relu'):
  model = Sequential()
  model.add(Dense(units= neurons,
                  activation= activation,
                  input_dim = 13,
                  name = 'Hidden1'))
  for i in range(hidden_layers-1):
    model.add(Dense(units= neurons,
                  activation= activation,
                  input_dim = neurons,
                  name = f'Hidden{i+2}'))

  model.add(Dense(units= 1,
                  activation= 'linear',
                  name = 'Output'))
  return model

model = create_model()
model.summary()


plot_model(model, show_layer_activations= True)
```

# Testing the 16 proposed neural networks 

```python
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import r2_score, mean_squared_error

es = EarlyStopping('val_loss', patience = 30, restore_best_weights= True)

layers = [1,2]
neurons = [10,20]
activations = ['relu', 'tanh']
lrs = [0.01,0.05]

results = {}

for layer in layers:
  for neuron in neurons:
    for activation in activations:
      for lr in lrs:
        r2_train = 0
        mse_train = 0
        r2_val = 0
        mse_val = 0
        r2_test = 0
        mse_test = 0

        n = 5

        for _ in range(n):
          model = create_model(layer, neuron, activation)
          model.compile(Adam(learning_rate= lr), 'mse')
          history = model.fit(X_train,y_train,validation_data= (X_val, y_val), epochs= 1000, callbacks= [es], verbose= 0)
          y_pred = model(X_train)
          r2_train += r2_score(y_train, y_pred)
          mse_train += mean_squared_error(y_train, y_pred)

          y_pred = model(X_val)
          r2_val += r2_score(y_val, y_pred)
          mse_val += mean_squared_error(y_val, y_pred)

          y_pred = model(X_test)
          r2_test += r2_score(y_test, y_pred)
          mse_test += mean_squared_error(y_test, y_pred)

          del model

        r2_train = r2_train/n
        mse_train = mse_train/n
        r2_val = r2_val/n
        mse_val = mse_val/n
        r2_test = r2_test/n
        mse_test = mse_test/n

        results[(layer, neuron, activation, lr)] = [r2_train, mse_train, r2_val, mse_val, r2_test, mse_test]
        print(results)
```

# Testing The best model with visualization 

```python
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import r2_score, mean_squared_error

es = EarlyStopping('val_loss', patience = 30, restore_best_weights= True)

model = create_model(2, 10, 'relu')
model.compile(Adam(learning_rate= 0.05), 'mse', metrics = [RSquare()])
history = model.fit(X_train,y_train,validation_data= (X_val, y_val), epochs= 1000, callbacks= [es], verbose= 1)
```

```python
y_pred = model(X_train)
r2_train = r2_score(y_train, y_pred)
mse_train = mean_squared_error(y_train, y_pred)

y_pred = model(X_val)
r2_val = r2_score(y_val, y_pred)
mse_val = mean_squared_error(y_val, y_pred)

y_pred = model(X_test)
r2_test = r2_score(y_test, y_pred)
mse_test = mean_squared_error(y_test, y_pred)

result = [r2_train, mse_train, r2_val, mse_val, r2_test, mse_test]
print(result)
```
